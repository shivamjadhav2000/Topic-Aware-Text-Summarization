```python
import nltk
import string
import re
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
#Removing special characters and convert to lowercase
def expand_contractions(text):
    if type(text) is str:
        for key in contractions:
            value = contractions[key]
            text = text.replace(key, value)
        return text
    else:
        return text
def clean_text(text):
    return re.sub(r'[^a-zA-Z\s]', '', text).lower()
# Removing stopwords
def stopwords_removal(text):
    words = text.split()
    filtered_words = []
    for word in words:
        if word not in stop_words:
            filtered_words.append(word)
    return ' '.join(filtered_words)
def remove_accented_chars(text):
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    return text
# Converting words to their base form
def lemmatize_words(text):
    words = text.split()
    lemmatized_words = []
    for word in words:
        lemmatized_words.append(lemmatizer.lemmatize(word))
    return ' '.join(lemmatized_words)


data_set['Cleaned_Questions'] = data_set['Question'].apply(expand_contractions)
data_set['Cleaned_Questions'] = data_set['Cleaned_Questions'].apply(clean_text)
data_set['Cleaned_Questions'] = data_set['Cleaned_Questions'].apply(stopwords_removal)
data_set['Cleaned_Questions'] = data_set['Cleaned_Questions'].apply(remove_accented_chars)
data_set['Cleaned_Questions'] = data_set['Cleaned_Questions'].apply(lemmatize_words)
data_set
```

```python
#Tokenization
nltk.download('punkt')
data_set['tokenized_words'] = data_set['Cleaned_Questions'].apply(word_tokenize)
data_set.head()
```

```python
data_set['text_for_vectorization'] = data_set['tokenized_words'].apply(lambda x: ' '.join(x))
data_set['text_for_vectorization']
```

```python
from itertools import count
#Vectorizing the data
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
#Initializing vectorizers
count_vectorizer = CountVectorizer(max_features = 5000)
tfidf_vectorizer = TfidfVectorizer(max_features = 5000)

# Applying CountVectorizer
count_matrix = count_vectorizer.fit_transform(data_set['text_for_vectorization'])
count_df = pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
count_df.head()


tfidf_matrix = tfidf_vectorizer.fit_transform(data_set['text_for_vectorization'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
tfidf_df.head()
```

#LDA

```python
from gensim.corpora import Dictionary
feature_names = tfidf_vectorizer.get_feature_names_out()
dictionary = Dictionary([feature_names])
```

```python
from sklearn.decomposition import TruncatedSVD
import numpy as np

n_components_svd = 100
svd = TruncatedSVD(n_components = n_components_svd, random_state = 42)
tfidf_reduced = svd.fit_transform(tfidf_df)

corpus_gensim = [[(i, freq) for i, freq in enumerate(doc) if freq > 0] for doc in tfidf_reduced]
```

```python
from gensim.models import LdaModel
num_topic = 25
lda_model = LdaModel(corpus = corpus_gensim, num_topics= num_topic, passes = 10, chunksize=1000, random_state=42)
for i, topic in lda_model.show_topics(formatted=True, num_topics=num_topic, num_words=10):
  print(f"Topic {i+1}: {topic}")
```
